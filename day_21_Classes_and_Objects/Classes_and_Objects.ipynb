```python
#Question 

'''
Create a class called PersonAccount. It has firstname, lastname, incomes, 
expenses properties and it has total_income, total_expense, account_info, add_income, 
add_expense and account_balance methods. 
Incomes is a set of incomes and its description. The same goes for expenses.
'''

class PersonAccount:
    def __init__(self, firstname, lastname):
        self.firstname = firstname
        self.lastname = lastname
        self.incomes = {}
        self.expenses = {}

    def add_income(self, amount, description):
        self.incomes[description] = amount

    def add_expense(self, amount, description):
        self.expenses[description] = amount

    def total_income(self):
        return sum(self.incomes.values())

    def total_expense(self):
        return sum(self.expenses.values())

    def account_balance(self):
        return self.total_income() - self.total_expense()

    def account_info(self):
        print(f"Account Information for {self.firstname} {self.lastname}:")
        print(f"Total Income: {self.total_income()}")
        print(f"Total Expense: {self.total_expense()}")
        print(f"Account Balance: {self.account_balance()}")
        print("\nIncome Details:")
        for description, amount in self.incomes.items():
            print(f"{description}: {amount}")
        print("\nExpense Details:")
        for description, amount in self.expenses.items():
            print(f"{description}: {amount}")


# Example usage:
person = PersonAccount("Abubakar", "Al-amin")

# Adding incomes and expenses
person.add_income(5000, "Salary")
person.add_income(1000, "Bonus")
person.add_expense(1500, "Rent")
person.add_expense(500, "Utilities")
person.add_expense(300, "Groceries")

# Displaying account information
person.account_info()
```
